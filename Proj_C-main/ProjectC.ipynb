```python
import sklearn.linear_model
import sklearn.pipeline

my_classifier_pipeline = sklearn.pipeline.Pipeline([
    #('my_bow_feature_extractor', CountVectorizer(min_df=1, max_df=1.0, ngram_range=(1,1))),
    ('my_classifier', sklearn.linear_model.LogisticRegression(random_state=100))#C=1.0, max_iter=20, random_state=101)),
])

my_parameter_grid_by_name = dict()
#my_parameter_grid_by_name['my_bow_feature_extractor__min_df'] = [1, 2, 4]
my_parameter_grid_by_name['my_classifier__C'] = np.logspace(-4, 4, 9)
my_parameter_grid_by_name['my_classifier__max_iter'] = [10, 50, 100, 200, 500]

my_scoring_metric_name = 'balanced_accuracy' # because of the raio of female/male in the y vector

grid_searcher = sklearn.model_selection.GridSearchCV(
    my_classifier_pipeline,
    my_parameter_grid_by_name,
    scoring=my_scoring_metric_name,
    cv=3,
    refit=False)

grid_searcher.fit(tr_u, tr_ys)
```
